## 7.3 相関係数と共分散の計算

### 7.3.1 corrメソッド

```python
import pandas as pd
import numpy as np

pd.set_option('display.notebook_repr_html', False)

adams_act_breath = pd.DataFrame(
                [
                  ['20230102', 92, 76],
                  ['20230103', np.nan, np.nan],
                  ['20230104', 65, 87],
                  ['20230105', 96, 76]
                ],
                columns=['Date', 'Act', 'Breath'])
print(adams_act_breath)
```

```python
# 相関係数を計算
adams_act_breath.corr()
```

```python
adams_act_breath.corr().plot()
```

### 7.3.2 corrwithメソッド

```python
adams_act = pd.DataFrame(
                [
                  ['20230102', 92],
                  ['20230103', np.nan],
                  ['20230104', 65],
                  ['20230105', 96]
                ],
                columns=['Date', 'Act'])

adams_breath = pd.DataFrame(
                [
                  ['20230102', 90],
                  ['20230103', 71],
                  ['20230104', 81],
                  ['20230105', 76]
                ],
                columns=['Date', 'Breath'])

print(adams_act)
```

```python
print(adams_breath)
```

```python
# 列名が異なるため、NaNが出力される
adams_act.corrwith(adams_breath)
```

```python
# adams_actのAct列とadams_breathのBreath列間の相関係数を計算
adams_act.corrwith(adams_breath['Breath'])
```

```python
# Breath列と他の列との相関係数を計算
adams_act_breath.corrwith(adams_act_breath['Breath'])
```

### 7.3.3 covメソッド

```python
adams_act_breath.cov()
```

```python
adams_act_breath.cov().plot()
```
