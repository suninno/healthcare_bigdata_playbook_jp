## 11.4 注意と転倒の相関関係

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.notebook_repr_html', False)

df_summary = pd.read_csv('fall_summary_cleaned.csv', encoding="shift_jis")
df_summary
```

```python
plt.rc('font', family='Yu Gothic')

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(1,1,1)
ax.scatter(df_summary['caution'], df_summary['fall'])
ax.set_xlabel('注意')
ax.set_ylabel('転倒')

plt.title('注意と転倒の相関関係')
plt.show()
```

## 11.5 注意と転倒の回帰分析を試す

```python
# 注意、転倒値抽出
X = df_summary.caution.values
Y = df_summary.fall.values

# X, Yを 2d arrayに変更
X = X.reshape(-1,1)
Y = Y.reshape(-1,1)
```

```python
# 回帰分析する
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(X, Y)

# r 係数
print(lr.coef_[0])
# y 切片
print(lr.intercept_)
```

```python
NewY = lr.coef_[0]*X +lr.intercept_

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(1,1,1)
ax.scatter(X, Y)
ax.set_xlabel('注意')
ax.set_ylabel('転倒')

plt.plot(X, NewY, color='red')
plt.title('y = {}*x + {}'.format(lr.coef_[0], lr.intercept_))
plt.show()
```

```python
#pip install statsmodels
import statsmodels.api as sm
pd.set_option('display.notebook_repr_html', False)

res = sm.OLS(Y, sm.add_constant(X)).fit()
res.summary()
```
