```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.notebook_repr_html', False)

# 不在情報を含めて見てみましょう
df_ratio = pd.read_csv('stacked_one_25.csv', dtype={'day':object, '不在':int, '未活動':int,'活動':int,'注意':int,'転倒':int}, encoding="shift_jis")
df_ratio.head()
```

```python
plt.style.use('default')
plt.rc('font', family='Yu Gothic')
plt.rcParams['figure.figsize'] = [10, 8]

day = df_ratio['day']
fall = df_ratio['転倒']
caution = df_ratio['注意']
action = df_ratio['活動']
non_action = df_ratio['未活動']

plt.bar(day, non_action, edgecolor='black', color='w',
        hatch="---", label='未活動')
plt.bar(day, action, bottom=non_action, edgecolor='black',
        color='0.9', hatch="...", label='活動')
plt.bar(day, caution, bottom=non_action+action, edgecolor='black', color='0.5', hatch="ooo", label='注意')
plt.bar(day, fall, bottom=non_action+action+caution, edgecolor='black', color='0', hatch='\\\\', label='転倒')

plt.legend(loc='best')
plt.xticks(rotation=45)
plt.show()
```
