## 9.2 ヒートマップグラフで可視化してみる

### 9.2.2. 可視化してみる

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.notebook_repr_html', False)

plt.rcParams['figure.figsize'] = [10, 8]
act = pd.read_csv('OTg6QzA_activities.csv')
act.shape #データの構造、行列のサイズ
```

```python
act.head(5) #先頭の５行のデータ表示
```

```python
X = act.X.to_numpy()
Y = act.Y.to_numpy()
```

```python
X # X値を表示
```

```python
Y # Y値を表示
```

```python
plt.style.use('grayscale')
plt.figure(facecolor='white')

plt.scatter(X, Y) # 散布図グラフを描く
plt.show()
```

```python
plt.figure(facecolor='white')
plt.hist2d(X, Y, bins=(25,25), cmap= plt.cm.Greys)
cb = plt.colorbar()
cb.set_label('frequency of appearance, bins=25')
```

```python
plt.figure(facecolor='white')
plt.hist2d(X, Y, bins=(40,40), cmap=plt.cm.Greys)
cb = plt.colorbar()
cb.set_label('frequency of appearance, bins=40')
```

```python
plt.figure(facecolor='white')

plt.hexbin(X, Y, gridsize=20, cmap=plt.cm.Greys)
cb = plt.colorbar()
cb.set_label('frequency of appearance, gridsize=20')
```

### 9.2.3 複数のグラフを作成してみる

```python
x1 = X[:16570]
x2 = X[16570:16570*2]
x3 = X[16570*2:16570*3]
x4 = X[16570*3:]

y1 = Y[:16570]
y2 = Y[16570:16570*2]
y3 = Y[16570*2:16570*3]
y4 = Y[16570*3:]
```

```python
plt.figure(facecolor='white')

# ４分の１
plt.subplot(2,2,1)
plt.hist2d(x1, y1, bins=20, cmap=plt.cm.Greys)
plt.title('1-Q')
plt.colorbar()

# ４分の2
plt.subplot(2,2,2)
plt.hist2d(x2, y2, bins=20, cmap=plt.cm.Greys)
plt.title('2-Q')
plt.colorbar()

# ４分の3
plt.subplot(2,2,3)
plt.hist2d(x3, y3, bins=20, cmap=plt.cm.Greys)
plt.title('3-Q')
plt.colorbar()

# ４分の4
plt.subplot(2,2,4)
plt.hist2d(x4, y4, bins=20, cmap=plt.cm.Greys)
plt.title('4-Q')
plt.colorbar()

plt.show()
```

```python
plt.figure(facecolor='white')

X_dash = X[16570:]
Y_dash = Y[16570:]
plt.hist2d(X_dash, Y_dash, bins=20, cmap=plt.cm.Greys)
cb = plt.colorbar()
cb.set_label('frequency of appearance, bins=20')
```
